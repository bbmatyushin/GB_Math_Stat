Другими словами, модель линейной регрессии на этих признаках должна давать наилучший результат. В качестве ответа запишите значение коэффициента детерминации для модели на этих признаках.

__Ответ__: 0.76342462

````python
import numpy as np
import math
import scipy.stats as st
from itertools import combinations
````

````python
# Дано:

x = np.array([[ 1.22401313,  2.30868478,  3.03636353,  2.69287214],
     [-0.18757272,  1.30337355,  5.12093014,  3.46363202],
     [-0.81094525,  1.82463398,  5.79686488,  1.86159445],
     [ 0.75129018,  2.67392052,  3.65529809,  1.66746094],
     [ 0.00972362,  1.97367255,  2.50594319,  1.69755173],
     [-0.62972637,  0.77750764,  2.84124027,  4.54410559],
     [ 2.29536229,  1.81206697,  1.95026215,  1.51874636],
     [ 0.0920418 ,  2.26971361,  7.47708735,  2.61081203],
     [ 2.39252799,  3.17563985,  3.61420599,  5.10773362],
     [ 0.54983815,  2.87988651,  1.65752765,  1.59635987]])

y = np.array([9.26193358,  9.700363,  8.67214805,  8.74796974,  6.18689108,
      7.53312713,  7.57643777, 12.44965478, 14.29010746,  6.68361218])
````

````python
def R2(x:np.ndarray, y:np.ndarray):
    
    # Добавим столбец из 1, соответствующий нулевому фактору
    ones = np.ones((x.shape[0],1))
    x = np.hstack([ones, x])
    
    XTX = x.T.dot(x)

    # Коэффициенты:
    XTX_inv = np.linalg.inv(XTX)    
    b = XTX_inv.dot(x.T).dot(y)
    
    z = x.dot(b)
    e = y - z
    
    return 1 - e.var() / y.var()
````

````python
kd=[]
for i, j in combinations(range(x.shape[1]), 2):
    r = R2(x[:, [i,j]], y)
    print(i, j, r)
    kd.append(r)

print("\nОтвет: ",round(max(kd),8))
````

Является ли значимым уравнение регрессии, полученное в предыдущей задаче? В качестве ответа запишите 1, если является, и 0 иначе.

__Ответ__: 1


````python
k1 = 2
k2 = x.shape[0]- 2 - 1
RR2 = R2(x[:, [0,2]], y)
F = (RR2 / k1) / ((1 - RR2) / k2)
alpha = 0.05
t = st.f.ppf(1 - alpha, k1, k2)
print("Ответ: ", int(F>=t))
````

__Задача 3__

Для проведения A/B-тестирования сайта интернет-магазина были получены следующие данные: страница `A` была посещена 2509 раз, из них 77 закончились совершением покупки, страница `B` была посещена 1465 раз, 60 из них закончились совершением покупки. Является ли значимым отличие конверсии на страницах `A` и `B`? В качестве ответа запишите 1, если является, и 0 иначе.

_Подсказка_. Реализуйте двухвыборочный t-тест. В качестве выборок здесь можно взять наборы меток совершения покупки (0 или 1) каждым посетителем.

__Ответ__: 0



````python
# Дано:

y1 = np.zeros(2509)
y1[np.arange(77)] = 1
y2 = np.zeros(1465)
y2[np.arange(60)] = 1
````

````python
y = np.concatenate([y1, y2])

n1 = y1.shape[0]
n2 = y2.shape[0]
n = n1+n2

y1_mean = y1.mean()
y2_mean = y2.mean()
y_mean = y.mean()
````

````python
SSb = n1*(y1_mean - y_mean)**2 + n2*(y2_mean - y_mean)**2
SSw = ((y1 - y1_mean)**2).sum() + ((y2 - y2_mean)**2).sum()
````

````python
k = 2
k1 = k-1
k2 = n-k
sigma2b = SSb/k1
sigma2w = SSw/k2

F = sigma2b/sigma2w
````

````python
alpha = 0.05
t = st.f.ppf(1 - alpha, k1, k2)
print("Ответ: ",int(F>=t))
````
